# Down sampling by taking the n'th element.

```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
```

```python
# Create the data
x0 = np.arange(0, 3, 0.1, dtype=float)
y0 = x0 * 2
df=pd.DataFrame({'x':x0, 'y':y0})
df
```

```python
# # Resampling
# pd.to_timedelta(df['x'], unit='S')

# df.time=pd.to_timedelta(df.time,unit="s")


df1 = df.copy()
# df1.x

df1.x = pd.to_timedelta(df1.x ,unit="s")
df1.set_index("x", inplace=True)

df1 = df1.resample("40ms").mean()

df1


# Here need to:
#   Interpolate the missing values.
#   Change 'x' back to float.
```
