**Analyzing Load Cell Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('_channel_lc1_raw.csv')
data = df.loc[(df['T[ms]'] >= 40000) & (df['T[ms]'] <= 50000)]
```

```python
x = data['T[ms]']
y = data['LC1[tick]']
plt.plot(x,y)
plt.show()
```

```python
np.mean(y)
```

```python
np.mean(y)
rms = np.sqrt(np.mean((y-np.mean(y))**2))
rms
```
